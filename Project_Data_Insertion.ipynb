```python
import csv
```

```python
with open('production_Sites.csv', mode ='r')as file:
    csvFile = csv.reader(file)
    next(csvFile)
    for lines in csvFile:
        site_ID = lines[0]
        street_Address = lines[1]
        city = lines[2]
        zip_Code = lines[3]
        site_Name = lines[4]
        
        sql_insert = """
        INSERT INTO production_Sites VALUES({val1}, "{val2}", "{val3}", {val4}, "{val5}");
        """.format(val1 = site_ID, val2 = street_Address, val3 = city, val4=zip_Code, val5 = site_Name)
        
        c.execute(sql_insert)

conn.commit()
```
